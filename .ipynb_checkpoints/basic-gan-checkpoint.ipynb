```python
import os
from itertools import cycle
from torch.utils.data import Dataset, DataLoader
from torchvision import datasets
from torchvision.io import read_image
from torchvision.transforms import Compose, Resize, Normalize, ToTensor
import torch
from torch import nn
import torch.optim as optim
import torchvision.utils as vutils
import itertools
from torch.nn import MSELoss, L1Loss
from torch.autograd import Variable
from torch import Tensor

from PIL import Image
import numpy as np

```

```python
class ResidualBlock(nn.Module):
    """Implements a single block in the ResNet architecture."""
    def __init__(self, in_features):
        super(ResidualBlock, self).__init__()

        self.block = nn.Sequential(
            nn.ReflectionPad2d(1),
            nn.Conv2d(in_features, in_features, 3),
            nn.InstanceNorm2d(in_features),
            nn.ReLU(inplace=True),
            nn.ReflectionPad2d(1),
            nn.Conv2d(in_features, in_features, 3),
            nn.InstanceNorm2d(in_features)
        )

    def forward(self, x):
        return x + self.block(x)
```

```python
class Generator(nn.Module):
    """Defines the architecture of the generator network for CycleGAN."""
    def __init__(self, input_channels=3, output_channels=3, num_residual_blocks=9):
        super(Generator, self).__init__()

        # Initial convolution block
        model = [
            nn.ReflectionPad2d(3),
            nn.Conv2d(input_channels, 64, kernel_size=7),
            nn.InstanceNorm2d(64),
            nn.ReLU(inplace=True)
        ]

        # Downsampling
        in_features = 64
        out_features = in_features*2
        for _ in range(2):
            model += [
                nn.Conv2d(in_features, out_features, kernel_size=3, stride=2, padding=1),
                nn.InstanceNorm2d(out_features),
                nn.ReLU(inplace=True)
            ]
            in_features = out_features
            out_features = in_features*2

        # Residual blocks
        for _ in range(num_residual_blocks):
            model += [ResidualBlock(in_features)]

        # Upsampling
        out_features = in_features//2
        for _ in range(2):
            model += [
                nn.ConvTranspose2d(in_features, out_features, kernel_size=3, stride=2, 
                                   padding=1, output_padding=1),
                nn.InstanceNorm2d(out_features),
                nn.ReLU(inplace=True)
            ]
            in_features = out_features
            out_features = in_features//2

        # Output layer
        model += [
            nn.ReflectionPad2d(3),
            nn.Conv2d(64, output_channels, kernel_size=7),
            nn.Tanh()
        ]

        self.model = nn.Sequential(*model)

    def forward(self, x):
        return self.model(x)
```

```python
class Discriminator(nn.Module):
    """Defines the architecture of the discriminator network for CycleGAN."""
    def __init__(self, input_channels=3):
        super(Discriminator, self).__init__()

        # A sequence of convolutions with increasing number of filters and stride 2
        # This reduces the spatial dimensions and increases depth with each layer
        model = [
            self.discriminator_block(input_channels, 64, normalization=False),
            self.discriminator_block(64, 128),
            self.discriminator_block(128, 256),
            self.discriminator_block(256, 512, stride=1)
        ]

        # The final convolutional layer that outputs a single dimensional output
        # for each patch classified
        model += [nn.Conv2d(512, 1, kernel_size=4, stride=1, padding=1)]

        self.model = nn.Sequential(*model)

    def discriminator_block(self, in_filters, out_filters, stride=2, normalization=True):
        """Returns layers of each discriminator block"""
        layers = [nn.Conv2d(in_filters, out_filters, kernel_size=4, stride=stride, padding=1)]
        if normalization:
            layers.append(nn.InstanceNorm2d(out_filters))
        layers.append(nn.LeakyReLU(0.2, inplace=True))
        return nn.Sequential(*layers)

    def forward(self, img):
        # img is a batch of images
        return self.model(img)
```

```python
transform = Compose([
    Resize((256, 256)),
    ToTensor(),
    Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

monet_dataset = datasets.ImageFolder('./training-data/monet_jpg', transform=transform)
regular_image_dataset = datasets.ImageFolder('training-data/photo_jpg', transform=transform)

dataloader_monet = DataLoader(monet_dataset, batch_size=4, shuffle=True, pin_memory=True)
dataloader_real  = DataLoader(regular_image_dataset, batch_size=4, shuffle=True, pin_memory=True)
```

```python
G = Generator()
F = Generator()
D_X = Discriminator()
D_Y = Discriminator()
epochs = 10
```

```python
# Define optimizers for each model
optimizer_G = torch.optim.Adam(itertools.chain(G.parameters(), F.parameters()), lr=0.0002, betas=(0.5, 0.999))
optimizer_D_X = torch.optim.Adam(D_X.parameters(), lr=0.0002, betas=(0.5, 0.999))
optimizer_D_Y = torch.optim.Adam(D_Y.parameters(), lr=0.0002, betas=(0.5, 0.999))
```

```python
# Loss functions
adversarial_loss = MSELoss()  # For GAN Loss
cycle_loss = L1Loss()  # For Cycle Consistency Loss
identity_loss = L1Loss()  # For Identity Loss (optional)
```

```python
# make sure cuda is being used 
if torch.cuda.is_available():
    print("CUDA is available. Training on GPU.")
    device = torch.device("cuda")
else:
    print("CUDA is not available. Training on CPU.")
    device = torch.device("cpu")


G = G.to(device)
F = F.to(device)
D_X = D_X.to(device)
D_Y = D_Y.to(device)
```

```python
def save_generated_images(epoch, generator, output_dir='generated_images', img_size=(256, 256)):
    
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    with torch.no_grad():
        
        generator.eval()  
        
        image_path = 'training-data/input-data/000ded5c41.jpg'

        image = Image.open(image_path).convert('RGB')
        input_image = transform(image).unsqueeze(0)  # Add batch dimension
        input_image = input_image.to(device)

        with torch.no_grad():  # No need to track gradients
            output_image = generator(input_image)
        
        generator.train()  

    # Normalize images to [0,1] and save
    filename = os.path.join(output_dir, f'epoch_{epoch}.png')
    vutils.save_image(output_image, filename, normalize=True)

    print(f'Saved generated images to {filename}')
```

```python
gan_loss = []
discriminator_loss = []
```

```python
lambda_cycle = 10.0  # Weight for cycle consistency loss
lambda_identity = 5.0  # Weight for identity loss
batch_size = 4
epoch_count = 1
for epoch in range(epochs):
    for x, y in zip(dataloader_monet, dataloader_real):
            
        #real_X = Variable(x[0].type(Tensor))  # Monet paintings
        #real_Y = Variable(y[0].type(Tensor))  # Real images

        real_X = x[0].to(device)
        real_Y = y[0].to(device)
        
        output_shape = [4, 1, 30, 30]

        #torch.Size([4, 1, 30, 30])
        
        # Adversarial ground truths
        # valid = Variable(Tensor(np.ones((real_X.size(0), *output_shape))), requires_grad=False)
        # fake = Variable(Tensor(np.zeros((real_X.size(0), *output_shape))), requires_grad=False)

        valid = torch.ones(output_shape, device=device)  # Assuming `batch_size` is defined
        fake = torch.zeros(output_shape, device=device)

        # ------------------
        #  Train Generators
        # ------------------

        optimizer_G.zero_grad()

        # Identity loss
        loss_id_X = identity_loss(G(real_X), real_X)
        loss_id_Y = identity_loss(F(real_Y), real_Y)

        # GAN loss
        fake_Y = G(real_X)
        loss_GAN_G = adversarial_loss(D_Y(fake_Y), valid)
        fake_X = F(real_Y)
        loss_GAN_F = adversarial_loss(D_X(fake_X), valid)

        # Cycle loss
        recovered_X = F(fake_Y)
        loss_cycle_X = cycle_loss(recovered_X, real_X)
        recovered_Y = G(fake_X)
        loss_cycle_Y = cycle_loss(recovered_Y, real_Y)

        # Total loss
        loss_G = loss_GAN_G + loss_GAN_F + lambda_cycle * (loss_cycle_X + loss_cycle_Y) + lambda_identity * (loss_id_X + loss_id_Y)
        loss_G.backward()
        optimizer_G.step()

        gan_loss.append(loss_G)

        # -----------------------
        #  Train Discriminator D_X
        # -----------------------

        optimizer_D_X.zero_grad()

        # Real loss
        loss_real = adversarial_loss(D_X(real_X), valid)
        # Fake loss (on batch of previously generated samples)
        loss_fake = adversarial_loss(D_X(fake_X.detach()), fake)
        # Total loss
        loss_D_X = (loss_real + loss_fake) / 2
        loss_D_X.backward()
        optimizer_D_X.step()

        # -----------------------
        #  Train Discriminator D_Y
        # -----------------------

        optimizer_D_Y.zero_grad()

        # Real loss
        loss_real = adversarial_loss(D_Y(real_Y), valid)
        # Fake loss (on batch of previously generated samples)
        loss_fake = adversarial_loss(D_Y(fake_Y.detach()), fake)
        # Total loss
        loss_D_Y = (loss_real + loss_fake) / 2
        discriminator_loss.append(loss_D_Y)
        loss_D_Y.backward()
        optimizer_D_Y.step()

    save_generated_images(epoch_count, G)
    print("epoch " + str(epoch_count))
        #print("real loss: ")
        ##print(loss_real)
        #print("fake loss: ")
        #print(loss_fake)
    epoch_count = epoch_count + 1
    
        


```
